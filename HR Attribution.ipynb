```python
# visualization
#Check Attrition Rate
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
print(attrition_rate)

plt.figure(figsize=(3,2))
sns.barplot(x=attrition_rate.index, y=attrition_rate.values, palette="coolwarm")
plt.title("Employee Attrition Rate (%)")
plt.ylabel("Percentage")
plt.show()
```

âœ… Insight: ~16% of employees left the company.

```python
avg_tenure = df['YearsAtCompany'].mean()
print(f"Average employee tenure: {avg_tenure:.2f} years")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,5))

sns.histplot(df['Age'], kde=True, ax=axes[0], color='teal')
axes[0].set_title("Age Distribution")

sns.countplot(data=df, x='Gender', ax=axes[1], palette='Set2')
axes[1].set_title("Gender Distribution")

sns.countplot(data=df, x='Department', ax=axes[2], palette='Set1')
axes[2].set_title("Department Distribution")

plt.tight_layout()
plt.show()

```

```python
# Attrition by Overtime
sns.countplot(data=df, x='OverTime', hue='Attrition', palette='Set2')
plt.title("Attrition by Overtime")
plt.show()

```

```python
# Attrition by Job Role
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='JobRole', hue='Attrition', palette='Set2')
plt.title("Attrition by Job Role")
plt.xticks(rotation=45)
plt.show()


```

```python
# Attrition by Work Life Balance
sns.countplot(data=df, x='WorkLifeBalance', hue='Attrition', palette='coolwarm')
plt.title("Attrition by Work Life Balance")
plt.show()
```

ðŸ’¡ Insights
- Overtime employees leave more.

 - Sales & R&D have higher turnover.

 - Poor work-life balance â†’ higher attrition.
